```python
def RestaurantSearch(City,Cuisine, min_price, max_price):
    zomato_data = df[(df['Cuisines'].apply(lambda x: Cuisine.lower() in x.lower())) &
                (df['City'].apply(lambda x: City.lower() in x.lower())) &
                (df['Average Cost for two']>=min_price) &
                (df['Average Cost for two']<=max_price)]

    zomato_data = df[['Restaurant Name','Address','Average Cost for two','Aggregate rating']]
    zomato_data = zomato_data.sort_values(by='Aggregate rating', ascending=False)
    return zomato_data
```

```python
z_data = RestaurantSearch('Bangalore','South Indian', 500,1000)
```

```python
ZomatoData  = df
```

```python
def RestaurantSearch(City,Cuisine, min_price, max_price, return_top=5):
    logging.info(f"Input City Name: {City}, Cuisine name: {Cuisine}, Price range: {min_price} to {max_price}")
    zomato_data = ZomatoData[(ZomatoData['Cuisines'].apply(lambda x: Cuisine.lower() in x.lower())) &
     (ZomatoData['City'].apply(lambda x: City.lower() in x.lower())) &
     (ZomatoData['Average Cost for two']>=min_price) &
     (ZomatoData['Average Cost for two']<=max_price)
     ]

    zomato_data = zomato_data[['Restaurant Name','Address','Average Cost for two','Aggregate rating']]
    zomato_data = zomato_data.sort_values(by='Aggregate rating', ascending=False)
    return zomato_data.iloc[:return_top]
```

```python
results = RestaurantSearch('Bangalore', 'North Indian', 500, 1500)
results
```

```python
def format_message(df):
    resp = ""
    for _ , restaurant in df.iterrows():
        r_temp = f" Restaurnat Name: {restaurant['Restaurant Name']}\n Address: {restaurant['Address']}\n Average cost for two: {restaurant['Average Cost for two']} \n Rating is {restaurant['Aggregate rating']} \n {'-'*30} \n"
        resp += r_temp
    return resp
```

```python
response = f"Here are the top 5 matches for your search: \n"
response += format_message(results)
print(response)
```

```python
import re
def parse_price(raw_price):
    low_price= 0
    high_price = 100000
    temp = re.findall(r'\d+', raw_price)
    res = list(map(int, temp))
    print(res)
    low_list = ['less', '<','within','low','']
    high_list = ['more','>', 'higher']
    # Logic to get the price
    if len(res)>1:
        low_price = res[0]
        high_price = res[-1]
    if len(res) == 1:
        b_flag = [ele for ele in high_list if(ele in raw_price)]
        if len(b_flag) > 0:    
            low_price = res[0]
        else:
            high_price = res[0]
    return low_price, high_price
```
